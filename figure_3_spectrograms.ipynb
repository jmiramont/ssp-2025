```python
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Audio
import scipy.signal as sg
import librosa
from src.utilities.utilstf import get_spectrogram
import os
from matplotlib import ticker
import seaborn as sns
import librosa

plt.rcParams['xtick.labelsize'] = 6.0
plt.rcParams['ytick.labelsize'] = 6.0
plt.rcParams['axes.labelsize'] =  6.0
```

```python
# plt.rc('text', usetex=True)
# plt.rc('font', family='serif')
```

```python
N = int(1*8192)
Nfft = 2*N
```

### Load audio signals

```python
signals_dict = {}
for i in range(6,7):
    x,fs = librosa.load('signals/{}_male.mp3'.format(i), sr=8000)
    # x,fs = librosa.load('signals/capev_male.wav', sr=8000)
    assert len(x)>N, 'Signal length not compatible.'
    x = x[:N]
    signals_dict['{}_male'.format(i)]=x

T = N/fs
    
Audio(x, rate=fs)
```

```python
fig, ax = plt.subplots(1,1, figsize=(1.4,1.4))
S, _ = get_spectrogram(signals_dict['6_male'],Nfft=Nfft)
ax.imshow(np.log(S+1e-6),origin='lower',aspect='auto', extent=[0,T,0,fs/2])

ax.set_xlim([0,1.055])
# Change axes to kHz and ms
import matplotlib.ticker as tkr     # has classes for tick-locating and -formatting
def numfmt(x, pos):
    s = '{}'.format(x / 1000.0)
    return s
yfmt = tkr.FuncFormatter(numfmt)    # create your custom formatter function
ax.yaxis.set_major_formatter(yfmt)

def numfmt(x, pos):
    s = '{}'.format(x)
    return s
yfmt = tkr.FuncFormatter(numfmt)    # create your custom formatter function
ax.xaxis.set_major_formatter(yfmt)

ax.set_xlabel(r'time (s)')
# ax.set_ylabel('frequency (kHz)',fontsize=7)
ax.set_title(r'Speech - Male', fontsize=7)
# ax.set_xticklabels(ax.get_xticklabels(),fontsize=)
ax.set_yticklabels([],fontsize=6)

ax.get_yaxis().set_visible(False)
sns.despine(offset=2.0,trim=False)
ax.spines['left'].set_visible(False)

S=[]
# Save figure
fig.savefig('figures/spectrogram_male.pdf',
             dpi=900, 
             transparent=False, 
             bbox_inches='tight', 
             pad_inches=0.01,
            )
```
